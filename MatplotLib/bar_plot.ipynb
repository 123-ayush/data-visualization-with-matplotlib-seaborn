Bar Charts:
- A bar plot (or bar chart) is a graph that displays categorical data using rectangular bars.
- Each bar's height (or length, in horizontal bars) represents a value.

- plt.text(x-axis pos, y-axis pos, text, alignment)
- plt.xlim(x_lower_limit, x_upper_limit)
- plt.ylim(y_lower_limit, y_upper_limit)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
oscar_movies = [
    "The dark knight",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 132] # in $M

plt.bar(years, oscar_revenue, color="green")
plt.title("Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+10, str(oscar_revenue[i]), ha="center")
```

Compare 2 sets of Data:
- We customise width. Width controls how thick each bar is. Usually ranges from 0 to 1. [Default val is 0.8]

```python
oscar_movies = [
    "The dark knight",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 132] # in $M

non_oscar_movies = [
    "Slumdog Millionare",
    "Avatar", 
    "Inception", 
    "Hugo", 
    "Lincoln"]

non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

x = np.arange(len(years))
width = 0.4

plt.title("Oscar vs Non-Oscar movies revenue")
plt.bar(x-width/2, oscar_revenue, width, label="Oscar movies")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Movies")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

# for i in range(len(years)):
#     plt.text(years[i], oscar_revenue[i]+10, str(oscar_revenue[i]), ha="center")
#     plt.text(years[i], non_oscar_revenue[i]+10, str(non_oscar_revenue[i]), ha="center")

plt.legend()
plt.xticks(x, years)
plt.show()
```
