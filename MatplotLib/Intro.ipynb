```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
Y = [5,6,7,8]

plt.plot(X, Y) #line plot
plt.show() # to show the plot
```

```python
# real data
oscar_movies = [
    "The dark knight",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 132] # in $M

plt.plot(years, oscar_revenue)
plt.title("Oscar movies revenue in each year.")
plt.xlabel("Oscar years")
plt.ylabel("Revenue in Million Dollars")
plt.show()
```

```python
non_oscar_movies = ["Slumdog Millionare", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, color="red", label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar winning Movies revenue vs Non oscar winning movies revenue")
plt.xlabel("Oscar Years")
plt.ylabel("Revenue (in $M)")

plt.legend(loc="upper left") # loc can be "upper left", "lower left", "upper right", "lower right", "center"
plt.grid()
plt.show()
```
